```python
import easyocr
import matplotlib.pyplot as plt
import cv2
from pylab import rcParams
from IPython.display import Image
import pandas as pd
```

```python
rcParams['figure.figsize']=8,16
reader=easyocr.Reader(['en','hi'])
```

```python
df=pd.read_csv("D://OCR//URL.csv")
l=[]
count=0
urls=df["URLS"]
for index,i in enumerate(urls):
    s=''
    file_name=f'{i}'
    output=reader.readtext(file_name)
    for j in output:
        s+=j[1]
    if len(s)!=0:
        l.append(s)
    else:
        l.append("NA")
    count+=1
    print(count)
df['Text']=l
df.to_csv("D://OCR//output.csv")
```
